# __Berlein et all 2003__


3 types: RS, FS, LTS (low-threshod spiking) in layer 4 of rat neocortex - in virto study.

    RS - regular spiking with srong adaptation; typically no fAHP

    FS-  fast spiking with no adaptation, high spiking frecquency around 100Hz, no adaptation;

    LTS - adaptation present, but lower than RS, small fAHP present
    
So RS from FS you can distinguih by looking on presence of adaptation, spiking frecquency and presence of fast AHP.
The LTS and RS cells you can distinguih by looking how their min AHP in between spikes is changing. If the difference between first and last through valur is negative it will be RS; if the difference will be positive it will be LTS

# Minimum plan for classifying cells based on the electrophysiology properties

Taking into accout the variability and number of possible classifications, especially for interneurons the first classification would be: Regular Spiking, Fast Spiking and other cells, whcih would be approximately a classification for piramidal and non piramidal cells (= pyramidal/interneuron/other). 
To classify cells into pyramidal/interneuron/other group I would use:
- adaptation presence, defined as the difference of mean and first ISI (so the stronger adaptation the more positive number) If yes:RS, no:FS some:other
- mean and first spike width ~1ms: RS ~0.5 FS in between:other
- presence of fAHP: yes if 5sec window fAHP < median PHP if yes: other or FS, no :RS
- fAHP amplitude /or spike treshold-PHP
- log ISI distribution/ by eye - if there are bursts 

These set of variables alows basic classification. Additionally, we could try to use other classification for interneurons further, as during recordings I encourted cells looking like chaterring/stuterring, bursting, delayed cells.
